```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_columns = [
    "Age_Group", "Gender", "Marital_Status", "Children",
    "Occupation", "Education_Level", "Location", "Region",
    "Income", "Product", "Policy_Type", "Policy_Status", "Claim_History"
]

# Set style
sns.set(style="whitegrid")

# count plots for all categorical columns
for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="Set2")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Save BEFORE showing
    plt.savefig(f"{col}_distribution.png", dpi=300, bbox_inches="tight")
    plt.show()
```
